```python
import pandas as pd
pd.options.mode.chained_assignment = None  # default='warn'
df = pd.read_csv("mydata.csv")
df.dtypes
```

```python
df.head()
```

```python
def convert_currency(val):
    """
    Convert the string number value to a float
     - Remove $
     - Remove commas
     - Convert to float type
    """
    new_val = val.replace(',','').replace('$', '')
    return float(new_val)
df['Amount'] = df['Amount'].apply(convert_currency)
df['Item Total']= df['Item Total'].apply(convert_currency)
df.head()
```

```python
df1 = df[['Full Name', 'Amount']]
df1.rename(columns = {'Amount' : 'Total Dollar Amount'}, inplace = True)
df1.groupby("Full Name").sum().sort_values("Total Dollar Amount", ascending = False).head(20)
```

```python
df1_count = df1.groupby("Full Name").count().sort_values("Total Dollar Amount", ascending = False)
df1_count.rename(columns = {'Total Dollar Amount': "Number of Transactions"}, inplace = True)
df1_count.head(20)
```

```python
df2 = df[['Vendor Name', 'Amount']]
df2.rename(columns = {'Amount' : 'Total Dollar Amount'}, inplace = True)
df2.groupby("Vendor Name").count().sort_values("Total Dollar Amount", ascending = False).head(20)
```

```python
df2_count = df2.groupby("Vendor Name").count().sort_values("Total Dollar Amount", ascending = False)
df1_count.rename(columns = {'Total Dollar Amount': "Number of Transactions"}, inplace = True)
df2_count.head(20)
```

```python
import calendar

df['CH Signoff Date'] = pd.to_datetime(df['CH Signoff Date'])


# You can format your date as you wish
df['CH Signoff Date'] = df['CH Signoff Date'].dt.strftime('%b-%Y')
df.head()
```

```python
df3 = df[['CH Signoff Date', 'Amount']]
df3.rename(columns = {'Amount' : 'Total Spend','CH Signoff Date': 'Month' }, inplace = True)
df3.groupby("Month").sum().sort_values("Total Spend", ascending = False)
```

```python
df4 = df[['Department Name','CH Signoff Date', 'Amount']]
df4.rename(columns = {'Amount' : 'Total Spend','CH Signoff Date': 'Month' }, inplace = True)
df4.groupby(['Department Name','Month']).sum().sort_values(by = ['Department Name','Month'], ascending=[True, False])
```

```python
df4 = df[['Department Name','CH Signoff Date', 'Amount']]
df4.rename(columns = {'Amount' : 'Total Spend','CH Signoff Date': 'Month' }, inplace = True)
df4.groupby(['Department Name']).sum().sort_values(by = ['Total Spend'], ascending= False).head(20)
```
